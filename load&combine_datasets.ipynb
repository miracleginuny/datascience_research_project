```python
import pandas as pd
```

```python
### Load and combine quarterly datasets for all years

### 2013

year_quarters_2013 = [(2013, 'Q1'), (2013, 'Q2'), (2013, 'Q3'), (2013, 'Q3')]
### Lists of file paths for drugs 2013
drug_file_paths_2013 = [
                        '../Datasets/faers_ascii/drug/DRUG13Q1.txt',
                        '../Datasets/faers_ascii/drug/DRUG13Q2.txt',
                        '../Datasets/faers_ascii/drug/DRUG13Q3.txt',
                        '../Datasets/faers_ascii/drug/DRUG13Q4.txt'
                        ]

# Initialise empty list to hold dataframe
drug_2013 = []

# Loop over file paths and year-quarter information
for file_path, (year, quarter) in zip(drug_file_paths_2013, year_quarters_2013):
    # read each file into the dataframe
    try:
        df = pd.read_csv(file_path, low_memory=False, delimiter='$', encoding='utf-8')
    except UnicodeDecodeError:
        try:
            df = pd.read_csv(file_path, low_memory=False, delimiter='$', encoding='latin1')
        except Exception as e:
            print(f"Error reading {file_path}: {e}")
            
    df['year'] = year
    df['quarter'] = quarter
    # append dataframe to the list
    drug_2013.append(df)

# Combine all quarterly drugs dataframes into a single year of 2013 dataframe
combined_drug_2013 = pd.concat(drug_2013, axis=0)


### Lists of file paths for reac 2013
reac_file_paths_2013 = [
                        '../Datasets/faers_ascii/reac/REAC13Q1.txt',
                        '../Datasets/faers_ascii/reac/REAC13Q2.txt',
                        '../Datasets/faers_ascii/reac/REAC13Q3.txt',
                        '../Datasets/faers_ascii/reac/REAC13Q4.txt'
                        ]

# Initialise empty list to hold dataframe
reac_2013 = []

# Loop over file paths for reac 2013
for file_path, (year, quarter) in zip(reac_file_paths_2013, year_quarters_2013):
    # read each file into the dataframe
    try:
        df = pd.read_csv(file_path, low_memory=False, delimiter='$', encoding='utf-8')
    except UnicodeDecodeError:
        try:
            df = pd.read_csv(file_path, low_memory=False, delimiter='$', encoding='latin1')
        except Exception as e:
            print(f"Error reading {file_path}: {e}")
            
    # add year and quarter columns 
    df['year'] = year
    df['quarter'] = quarter
    # append dataframe to the list
    reac_2013.append(df)

# Combine all quarterly reac dataframes into a single year of 2013 dataframe
combined_reac_2013 = pd.concat(reac_2013, axis=0)

### Lists of file paths for demo 2013
demo_file_paths_2013 = [
                        '../Datasets/faers_ascii/demo/DEMO13Q1.txt',
                        '../Datasets/faers_ascii/demo/DEMO13Q2.txt',
                        '../Datasets/faers_ascii/demo/DEMO13Q3.txt',
                        '../Datasets/faers_ascii/demo/DEMO13Q4.txt'
                        ]

# Initialise empty list to hold dataframe
demo_2013 = []

# Loop over file paths and year-quarter information
for file_path, (year, quarter) in zip(demo_file_paths_2013, year_quarters_2013):
    # read each file into the dataframe
    try:
        df = pd.read_csv(file_path, low_memory=False, delimiter='$', encoding='utf-8')
    except UnicodeDecodeError:
        try:
            df = pd.read_csv(file_path, low_memory=False, delimiter='$', encoding='latin1')
        except Exception as e:
            print(f"Error reading {file_path}: {e}")
            
    df['year'] = year
    df['quarter'] = quarter
    # append dataframe to the list
    demo_2013.append(df)

# Combine all quarterly drugs dataframes into a single year of 2013 dataframe
combined_demo_2013 = pd.concat(demo_2013, axis=0)

# save dataframes into csv files
combined_drug_2013.to_csv('../Datasets/Yearly datasets/drug_2013.csv', index=False)
combined_reac_2013.to_csv('../Datasets/Yearly datasets/reac_2013.csv', index=False)
combined_demo_2013.to_csv('../Datasets/Yearly datasets/demo_2013.csv', index=False)
```
